```python
#from matplotlib.image import imread
#import cv2
import numpy as np
from PIL import Image,ImageTk
```

```python

###########################################################
####### Loading the Data set with Tkinter & Pandas#########
###########################################################
# Loading the data set
# The GUI to get file name
def read_file():
    import tkinter                     #GUI package
    from tkinter.filedialog import askopenfilenames    #For selection of file(dataset) in windows explorer

    root=tkinter.Tk()         #Creating Object for TK interface      
    filename = askopenfilenames(title="Please select the csv file of the data")   #Browsing & selecting the required file
    print(filename)                                                              #Prints the location with file name
    root.destroy()                                                               #Closing/destroy object of TK interface
    return filename
```

```python
import tkinter
from tkinter import messagebox
top = tkinter.Tk()

top.title("Image Recommendations") 

b_img=Image.open("C:/Users/mdnau/Pictures/nn.jpg")
photo_img=ImageTk.PhotoImage(b_img)

def one_img():
    filename=read_file()
    import glob
    global img_one
    img_one=[]
    image= Image.open(filename[0])
    img_one.append(np.array(image))
    print(np.shape(img_one))

def images():
    filename=read_file()
    import glob
    n=len(filename)
    global img_all
    img_all=[]
    for i in range(0,n):
        image= Image.open(filename[i])
        img_all.append(np.array(image))
    for k in range(0,n):
        print(np.shape(img_all[k]))

def knn():
    global distance
    distance=[]
    for i in range(0,len(img_all)):
    #print(images_list[array])
        eucli=np.sqrt(np.sum((img_one[0]-img_all[i])**2))
        distance.append([eucli,img_all[i]])
    distance=sorted(distance, key=lambda x: x[0])
    print(distance) 

def r():
    from PIL import Image
    for i in range(0,3):                    #Set how many pics you want to display  0,len(distanc) for all img
        o=Image.fromarray(distance[i][1],'RGB')
        print(Image.fromarray(distance[i][1],'RGB'))
        o.save('my.png')
        o.show()



B = tkinter.Button(top, text ="One Image", command = one_img,height=5,width=15,font=20)
C= tkinter.Button(top,text="Browse images", command=images,height=5,width=15,font=20)
D= tkinter.Button(top,text="Calculate", command=knn,height=5,width=15,font=20)
E= tkinter.Button(top,text="Recommendations",command=r,height=5,width=15,font=20)
label=tkinter.Label(top,image=photo_img)
label.config(height=180,width=180)
label.place(x=10000,y=10000)

label.pack()
B.pack()
C.pack()
D.pack()
E.pack()
top.mainloop()
```
