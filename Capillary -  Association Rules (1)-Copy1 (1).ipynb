### Jaccard's Distance between Users : User- User Similarity meaure  J(A,B)=(A intesection B) / (A union B)

```python
df2=np.array(np.matrix(df1))
    
```

```python
jaccard=[]
for i in range(len(df2)):
    for j in range(len(df2)):
        intsec=np.dot(df2[i],df2[j])
        union=np.logical_or(df2[i],df2[j])
        jacc=intsec/union
        jaccard.append(jacc)
        
```

```python
# Get the most frequent itemsets. Here we are considering the itemsets as frequent if it crosses
# the minimum threshold support
freq_itemset = apriori(df=df1, min_support=0.00001,max_len=2 ,use_colnames=True)
```

```python
#Question 1.) Which Item category was most frequent?
freq_itemset.sort_values(by=['support'],ascending=0)
```

```python
rules = association_rules(freq_itemset, min_threshold=.1) # By default the metric is confidence
```

```python
rules
```

```python
rules=rules.sort_values(by=['antecedents','support','consequent support'],ascending=[0,0,0])
```

```python
rules['flag'] = (rules.antecedents != rules.antecedents.shift()).astype(int)
```

```python
rules1=rules.loc[rules['flag']==1,:]
```
